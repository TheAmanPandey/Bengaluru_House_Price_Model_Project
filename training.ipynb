```python
def convertRange(x):
    temp=x.split('-')
    if(temp==2):
        return (float(temp[0])+float(temp[1]))/2
    try:
        return(float(x))
    except:
        return None
```

```python
def remove_outliers_sqft(df):
    df_output=pd.DataFrame()
    for key,subdef in df.groupby('location'):
        m=np.mean(subdef.price_per_sqft)
        st=np.std(subdef.price_per_sqft)
        gendef=subdef[(subdef.price_per_sqft>(m-st)) & (subdef.price_per_sqft<(m+st))]
        df_output=pd.concat([df_output,gendef],ignore_index=True)
    return df_output
data=remove_outliers_sqft(data)
data.describe()
```

```python
def remove_outliers_bhk(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
data=remove_outliers_bhk(data)
```

```python
X=data.drop(columns=['price'])
Y=data['price']
```

**Applying Linear Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.metrics import r2_score
```

```python
column_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
X_transformed=column_trans.fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_transformed,Y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
print("R2 Score:",r2_score(Y_test,Y_pred))
```
